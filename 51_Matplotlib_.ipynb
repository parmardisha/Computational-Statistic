```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function to plot
def f(x):
  return x**2  # Example: x squared

# Generate x values
x = np.linspace(-5, 5, 100)  # 100 points between -5 and 5

# Calculate corresponding y values
y = f(x)

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x) = x^2")

# Display the plot
plt.show()
```
